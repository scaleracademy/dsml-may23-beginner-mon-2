```python
top3_publishers = data["Publisher"].value_counts()[:3]
```

```python
top3_publishers
```

```python
'''
HOMEWORK ->
Determine the top 3 publishers according to historical global sales.
'''
```

```python
top3_data = data.loc[data["Publisher"].isin(top3_publishers.index)]
```

```python
top3_publishers.index
```

```python
top3_data
```

```python
sns.countplot(x = "Publisher", hue = "Genre", data = top3_data)
plt.legend(loc = (-0.5, 0))
```

```python
# STACKED VERSION OF THE SAME.
```

```python
sns.boxplot(
    x = "Publisher",
    y = "Global_Sales",
    data = top3_data
)
```

```python
fig = plt.figure(figsize = (15, 10))

plt.subplot(2, 3, 1) # -> 2 rows, 3 columns, 1st slot
sns.scatterplot(x = "NA_Sales", y = "EU_Sales", data = top3_data)

plt.subplot(2, 3, 6)
sns.scatterplot(x = "NA_Sales", y = "JP_Sales", data = top3_data)
```

```python
fig, ax = plt.subplots(2, 2, figsize = (15, 10))
ax[0,0].scatter(top3_data["NA_Sales"], top3_data["EU_Sales"])
ax[0,1].scatter(top3_data["NA_Sales"], top3_data["JP_Sales"])
ax[1,0].scatter(top3_data["NA_Sales"], top3_data["Other_Sales"])
ax[1,1].scatter(top3_data["NA_Sales"], top3_data["Global_Sales"])
```
