```python
movies = pd.read_csv("movies.csv", index_col = 0)
directors = pd.read_csv("directors.csv", index_col = 0)
```

```python
movies
```

```python
directors
```

```python
data = movies.merge(
    directors,
    how = "left",
    left_on = "director_id", 
    right_on = "id"
)
```

```python
data
```

```python
data.drop(["director_id", "id_y"], inplace = True, axis = 1)
```

```python
data
```

```python
# Which director is the most productive?
```

```python
data.groupby(["director_name"])["title"].count()
```

```python
data.groupby(["director_name"])["title"].count().sort_values(ascending = False)
```

```python
data_agg = data.groupby(["director_name"])[["year", "title"]].aggregate({
    "year": ["min", "max"],
    "title": "count"
})
```

```python
data_agg
```

```python
data_agg.columns
```

```python
data_agg["year"]
```

```python
data_agg
```

```python
data_final = data.groupby("director_name")[["year", "title"]].aggregate(
    year_max = ("year", "max"), # new_col = ("col", "aggregator")
    year_min = ("year", "min"),
    title_count = ("title", "count")
)
```

```python
data_final
```

```python
data_final.reset_index(inplace = True)
```

```python
data_final
```

```python
data_final["years_active"] = data_final["year_max"] - data_final["year_min"]
```

```python
data_final
```

```python
data_final["movies_per_year"] = data_final["title_count"] / data_final["years_active"]
```

```python
data_final
```

```python
data_final.sort_values("movies_per_year", ascending = False)
```

# Pfizer Dataset

```python
data = pd.read_csv("Pfizer_1.csv")
```

```python
data
```

```python
data.info()
```

```python
pd.melt(data, id_vars = ["Date", "Parameter", "Drug_Name"])

# pd.melt(dataset, id_vars = [list of all the columns you do not want to melt])
```

```python
data_melt = pd.melt(
    data,
    id_vars = ["Date", "Parameter", "Drug_Name"],
    var_name = "Time",
    value_name = "Reading"
)
```

```python
data_melt
```

```python
# Melt -> Convert a bunch of columns to rows
# Pivot -> Does the opposite!
```

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
)
```

```python
data_melt.pivot(
    index = ["Date", "Drug_Name", "Parameter"],
    columns = "Time",
    values = "Reading"
).reset_index()
```

```python
data_melt
```

```python
data_tidy = data_melt.pivot(
    index = ["Date", "Time", "Drug_Name"],
    columns = "Parameter",
    values = "Reading"
)
```

```python
data_tidy
```

```python
data_tidy = data_tidy.reset_index()
```

```python
data_tidy
```

```python
data_tidy.columns.name = None
```

```python
data_tidy.head()
```

```python
# Date wise mean value of Temperature for every drug
```

```python
data_tidy.pivot(
    index = ["Drug_Name"],
    columns = "Date",
    values = "Temperature"
)
```

```python
pd.pivot_table(
    data_tidy,
    index = "Drug_Name",
    columns = "Date",
    values = ["Temperature"],
    aggfunc = np.mean
)
```

```python
pd.pivot_table(
    data_tidy,
    index = "Drug_Name", # Rows
    columns = "Date", # columns
    values = ["Temperature", "Pressure"], # columns to be aggregated for values
    aggfunc = np.min # aggregator function
)
```
